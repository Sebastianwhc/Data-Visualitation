## Información sobre el DataSet

El DataSet `MobPrice` contiene información sobre características, especificaciones, etc., del teléfono móvil y su rango de precios, los cuales tienen valores del 0 al 3.

Las características de los datos son las siguientes:
*   `battery_power`: Potencia de la batería en mAh
*   `blue`: Tiene BlueTooth o no
*   `clock_speed`: Velocidad de reloj del microprocesador
*   `dual_sim`: El teléfono tiene soporte dual sim o no
*   `fc`: Megapíxeles de la cámara frontal
*   `four_g`: Tiene soporte 4G o no
*   `int_memory`: Memoria Interna en GigaBytes
*   `m_dep`: Fondo Móvil en centímetros (cm)
*   `mobile_wt`: Peso del teléfono móvil
*   `n_cores`: Número de núcleos en el procesador
*   `pc`: Megapíxeles de la cámara principal
*   `px_height`: Altura de resolución de píxeles
*   `px_width`: Ancho de resolución de píxeles
*   `ram`: RAM en MB
*   `sc_h`: Altura de la pantalla del móvil en cm
*   `sc_w`: Ancho de la pantalla del móvil en cm
*   `talk_time`: Mayor tiempo después de una sola carga
*   `three_g`: 3G o no
*   `touch_screen`: Tiene pantalla táctil o no
*   `wifi`: Tiene wifi o no
*   `price_range`: Categoría de predicición

### Parte 1: uso de Matplotlib

```python
# Gráfico de barras.
# Cálculo de frecuencias para cada categoría de precio
price_counts = df["price_range"].value_counts()

# Creación del gráfico de barras
fig, ax = plt.subplots(figsize=(10, 8))
ax.bar(price_counts.index, price_counts.values)

# Añadir etiquetas y título
ax.set_xlabel("Categoría de precio")
ax.set_ylabel("Número de teléfonos móviles")
ax.set_title("Distribución de teléfonos móviles por categoría de precio")

plt.show()
```

### Parte2: uso de Seaborn

```python
# Mapa de calor.
import pandas as pd
import seaborn as sns

# Creación de la matriz de correlación
corr_matrix = df.corr()

# Creación del mapa de calor
fig, ax = plt.subplots(figsize=(20, 15))
sns.heatmap(corr_matrix, annot=True, cmap="YlGnBu")

# Añadir título
ax.set_title("Mapa de calor de correlaciones entre características y precio")

plt.show()
```
