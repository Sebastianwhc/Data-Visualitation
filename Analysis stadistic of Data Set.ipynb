```python
# Creación de nuevo DataSet de Pandas únicamente con las variables numéricas.
df_new = df.select_dtypes(include=['int64','float64'])
df_new.head()
```

```python



# Mapa de calor.
corr_matriz = df_new.corr()
fig, ax = plt.subplots(figsize=(20, 15))
sns.heatmap(corr_matriz,annot=True,cmap="YlGnBu")
ax.set_title("Mapa de calor de matriz de correlación de variables")
```
