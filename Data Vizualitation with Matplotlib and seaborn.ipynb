## Función hist

La función hist() se utiliza para crear un histograma, que es una representación visual de la distribución de una variable numérica. Los parámetros más comunes de la función son bins y histtype.

### Ejemplo:


```python
import pandas as pd
import matplotlib.pyplot as plt


#Para cargar el Dataset desde el drive
from google.colab import drive
drive.mount('/content/drive')
file_path = '/content/drive/MyDrive/titanic-train.csv'
titanic = pd.read_csv(file_path)


plt.hist(titanic["Age"], bins=10, histtype="bar")


plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.title("Distribución de la edad de los pasajeros del Titanic")


plt.show()
```

## Función boxplot() :     

La función boxplot() se utiliza para crear un diagrama de caja, que es una representación visual de la distribución de una variable numérica.


## Ejemplo:

```python
plt.boxplot(titanic['Fare'], vert=False, patch_artist=True, notch=True, showfliers=True)
plt.xlabel('Tarifa')
plt.title('Boxplot de Tarifas')
plt.show()
```

## Funcion sccater():

La función scatter() en matplotlib se utiliza para crear un gráfico de dispersión, que se utiliza para mostrar la relación entre dos variables continuas. En un gráfico de dispersión, cada punto representa una observación en el conjunto de datos y su posición en el eje x e y representa los valores de las dos variables.

##Ejemplo:

```python
plt.scatter(titanic['Age'], titanic['Fare'], c=titanic['Survived'], marker='o', alpha=0.5)
plt.xlabel('Edad')
plt.ylabel('Tarifa')
plt.title('Diagrama de dispersión de Edad y Tarifa')
plt.show()
```

Del dataset Titanic programar código para obtener las siguientes gráficas utilizando las diferentes librerías para graficar matplolib o seaborn vistas en clase.

Como puede observar, las imágenes no tienen ni etiquetas en los ejes verticales, ni títulos. Deberá colocarlos haciendo referencia a la información correspondiente.

```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic_gender = titanic[titanic['Sex'].isin(['male', 'female'])]
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 6))
sns.countplot(data=titanic_gender, x='Sex', ax=ax1)
for p in ax1.patches:
    ax1.annotate(p.get_height(), (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom')
ax1.set_title('Total de pasajeros a bordo')
ax1.set_xlabel('Género')
ax1.set_ylabel('Cantidad')
sns.countplot(data=titanic_gender, x='Sex', hue='Survived', ax=ax2)
for p in ax2.patches:
    ax2.annotate(p.get_height(), (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom')
ax2.set_title('Total de sobrevivientes')
ax2.set_xlabel('Género')
ax2.set_ylabel('Cantidad')
ax2.legend(title='Sobreviviente', loc='upper right')
plt.show()
```
